```python
#filling the null values
df1 = df.fillna(value=0)
df1
```

```python
df2 = df.fillna(value=1)
df2
```

```python
#fillnull with  previous value
df3 = df.fillna(method='pad')
df3
```

```python
# filling null values with next upcoming value rowwise
df4 = df.fillna(method='bfill')
df4
```

```python
# previous column wise
df5 = df.fillna(method='pad',axis=1)
df5
```

```python
#  next upcoming column wise
df6 = df.fillna(method='bfill',axis=0)
df6
```

```python
# filing different values in null in different columns
df7 = df.fillna({'Revenue':'abcd',
               'Growth' :'pqrs'})
df7
```

```python
#filling null value with the min of column
df8 = df.fillna(value = df['Profit'].mean())
df8
```

```python
#dropna
df9 = df.dropna()
df9
```

```python
df10 = df.dropna(how='all')
df10
```

```python
df11 = df.dropna(how='any')
df11
```

```python
#replace
import numpy as np
df12 = df.replace(to_replace=np.nan,value=12345)
df12
```

```python
df13= df.replace(to_replace='7%',value='0%')
df13
```
